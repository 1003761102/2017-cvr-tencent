```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dtype = {'label':object, 'weekday':object, 'clickTime':object, 'creativeID':object, 'userID':object, 'positionID':object, 'connectionType':object,
       'telecomsOperator':object, 'age':object, 'gender':object, 'education':object, 'marriageStatus':object,
       'haveBaby':object, 'hometown':object, 'residence':object, 'sitesetID':object, 'positionType':object,
       'adID':object, 'camgaignID':object, 'advertiserID':object, 'appID':object, 'appPlatform':object,
       'appCategory':object}
test = pd.read_csv('../../data/test.csv', dtype=dtype)
```

```python
test.describe(include=['object'])
```

```python
def time_transform(series):
    return str(series)[2:4]
def convert_age(age):
    return int(int(age) / 5)
test['clickTime'] = test['clickTime'].apply(time_transform)
test['age'] = test['age'].apply(convert_age)
test['age'] = test['age'].astype('object')
```

```python
test.drop('userID', axis=1, inplace=True)
```

```python
# 画图查看分类变量的分布
categorical_feature = test.dtypes[test.dtypes == 'object'].index
for feature in categorical_feature:
    sns.countplot(x=test[feature], data=test)
    plt.show()
```
